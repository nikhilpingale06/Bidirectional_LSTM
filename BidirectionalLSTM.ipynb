```python
import pandas as pd
df = pd.read_csv("train.csv",engine='python',error_bad_lines = False)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df = df.dropna()
```

```python
df.shape
```

```python
#Dividing the dataset into dependent and independent features
X = df.drop('label',axis=1)
y = df['label']
```

```python
y.value_counts()
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Bidirectional
```

```python
#iNITIALIZING VOCABULARY SIZE
voc_size = 5000
```

```python
messages = X.copy()
```

```python
messages.head()
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
messages.reset_index(inplace=True)
```

```python
#Data preprocessing (Stemming and applying stopwords)
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []
for i in range(0,len(messages)):
  print(i)
  review = re.sub('[^a-zA-z]',' ',messages['title'][i])
  review = review.lower()
  review = review.split()

  review = [ps.stem(word) for word in review if word not in stopwords.words('english')]
  review = ' '.join(review)
  corpus.append(review)
```

```python
corpus
```

```python
onehot_repr = [one_hot(words,voc_size) for words in corpus]
onehot_repr
```

# Embedding Representation

```python
#Initializing the sentence length for padding
sent_length = 20
embedded_docs = pad_sequences(onehot_repr,padding='pre',maxlen= sent_length)

embedded_docs
```

```python
#Creating the model
embedding_vector_features = 40
model = Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(Bidirectional(LSTM(100)))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
```

```python
#converting embedded docs to array so we can supply it to model
import numpy as np
X_final = np.array(embedded_docs)
y_final = np.array(y)
```

```python
#Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_final,y_final,test_size=0.2,random_state=42)
```

```python
#Training
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=8)
```

```python
y_pred=model.predict(X_test)
```

```python
import numpy as np

y_pred=np.where(y_pred>=0.5,1,0)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
confusion_matrix(y_test,y_pred)
```

```python
print(accuracy_score(y_test,y_pred))
```
